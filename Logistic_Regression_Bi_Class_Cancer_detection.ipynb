```python
training_data = raw_data.iloc[0:int(0.7*raw_data.shape[0])]
```

```python
testing_data = raw_data.iloc[int(0.7*raw_data.shape[0]):]
```

```python
training_labels = labels.iloc[0:int(0.7*raw_data.shape[0])]
```

```python
C = np.array(training_labels).reshape(training_data.shape[0],1)
```

```python
testing_labels = labels.iloc[int(0.7*raw_data.shape[0]):]
```

```python
training_data = np.array(training_data)
```

```python
def thet0_plus_thet_tranpose_x(training_data,thet0,thet):   
    
    return (np.matmul(training_data,thet) + thet0)
```

```python
def sigmoid(score):
    
    return (1/(1 + np.exp(-score)))
```

```python
def neg_log_loss(H):
    
    return -((np.matmul(C.T,np.log(H)) + np.matmul((1-C).T,np.log(1-H)))/training_data.shape[0])
```

```python
mean_training_data=training_data.mean()
std_training_data=training_data.std()
```

```python
normalized_training_data=(training_data-mean_training_data)/std_training_data
```

```python
thet0_initial = np.random.randn(1,1)

thet_initial = np.random.randn(training_data.shape[1],1)

epsilon = 0.000001

learning_rate = 0.00001

iterations = []

neg_log_likeli_history = []

i = 0

while(True):
    
    
    score_initial = thet0_plus_thet_tranpose_x(normalized_training_data,thet0_initial,thet_initial)
    H_i_initial = sigmoid(score_initial)
    
    
    thet0_final = thet0_initial - (learning_rate * np.sum(H_i_initial - C))
    thet_final = thet_initial - (learning_rate * np.matmul(normalized_training_data.T,(H_i_initial-C)))
    
    
    score_final = thet0_plus_thet_tranpose_x(normalized_training_data,thet0_final,thet_final)
    H_i_final = sigmoid(score_final)
    
    
    if abs(neg_log_loss(H_i_initial) - neg_log_loss(H_i_final)) < epsilon:
        
        break
        
    print("Iteration # =",i,"and Neg Log Loss =",neg_log_loss(H_i_initial))
    
    iterations.append(i)
    neg_log_likeli_history.append(neg_log_loss(H_i_initial))
    i += 1
    
    
    thet0_initial = thet0_final
    thet_initial = thet_final    
```

True Positive(TP)=  Cancer was Malignant, Test classified it as Malignant.<br>
True Negative(TN)=  Cancer was Benign, Test classified it as Benign.<br>
False Positive(FP)= Cancer was Malignant, Test classified it as Benign.<br>
False Negative(FN)= Cancer was Benign, Test classified it as Malignant.<br>

<b> True positive rate= $\frac{True Positive}{True Positive + False Negative}$ </b> <br>
<br>
<b> False positive rate = $\frac{False  Positive} {False  Positive +True  Negative} $ </b>

```python
tp=0
tn=0
fp=0
fn=0

for i in range (C_testing.shape[0]):
    if boolean_mask[i]==True: # means our algorithm predicted that cancer is malignant
        if C_testing[i]==1: # cancer actually is malignant
            tp+=1  #predicted class=1,actual class=1, hence true positive
        else:
            fp+=1    #predicted class=1,actual class=0, hence 
            
    elif boolean_mask[i]==False: #  means our algorithm predicted that cancer is benign 
        if C_testing[i]==1:  #checking for cancer is malignant
            fn+=1            
        else:
            tn+=1
    i+=1
            
        
```

```python
c=[(tp,fp),(fn,tn)]
```

```python
c
```

```python
confusion_matrix=pd.DataFrame(c,index=['Predicted- Malignant','Predicted-Benign'],columns=['Actually-Malignant','Actually-Benign'])
```

```python
confusion_matrix
```

```python
tpr=(tp/(tp+fn))
fpr=(fp/(fp+tn))
```

```python
plt.plot([0,fpr,1],[0,tpr,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

```python
#AUC ROC
score=np.trapz([0,tpr,1],[0,fpr,1])
```

The AUC-ROC score tells us the robustness of our model i.e. higher the score better our model performs on unseen data.
